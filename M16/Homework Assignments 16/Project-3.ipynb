```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("Bestsellers with categories.csv")
```

```python
data.info()
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
q1 = np.quantile(data['Price'], 0.25)
q2 = np.quantile(data['Price'], 0.50)
q3 = np.quantile(data['Price'], 0.75)
```

```python
print("First Quartile - ", q1)
```

```python
print("Second Quartile - ", q2)
```

```python
print("Third Quartile - ", q3)
```

```python
IQR_price = q3 - q1
print("Interquartile Range - ", IQR_price)
```

```python
q1 = np.quantile(data['User Rating'], 0.25)
q2 = np.quantile(data['User Rating'], 0.50)
q3 = np.quantile(data['User Rating'], 0.75)
```

```python
print("First Quartile - ", q1)
```

```python
print("Second Quartile - ", q2)
```

```python
print("Third Quartile - ", q3)
```

```python
IQR_rating = q3 - q1
print("Interquartile Range - ", IQR_rating)
```
