```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
PenguinDF = pd.read_csv("Penguins Data.csv")
```

```python
PenguinDF = PenguinDF[
    ["Culmen Length (mm)", "Culmen Depth (mm)", "Flipper Length (mm)", "Body Mass (g)"]
].dropna()
```

```python
sns.scatterplot(
    data=PenguinDF,
    x="Culmen Length (mm)",
    y="Body Mass (g)"
)
plt.show()
```

```python
sns.scatterplot(
    data=PenguinDF,
    x="Culmen Depth (mm)",
    y="Body Mass (g)"
)
plt.show()
```

```python
sns.pairplot(PenguinDF)
plt.show()
```

```python
sorted_df = PenguinDF.sort_values("Culmen Length (mm)")

plt.fill_between(
    sorted_df["Culmen Length (mm)"],
    sorted_df["Body Mass (g)"]
)
plt.show()
```
