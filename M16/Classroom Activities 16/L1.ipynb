```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Titanic Dataset.csv")
```

```python
data.isnull().sum()
```

```python
median_age = np.median(data['Age'])
print("Median of Age is : ", median_age)
```

```python
median_fare = np.median(data['Fare'])
print("Median of Fare is : ", median_fare)
```

```python
mode_age = stats.mode(data['Age'])
print("Mode of Age is : ", mode_age)
```

```python
mode_class = stats.mode(data['Pclass'])
print("Mode of PClass is : ", mode_class)
```

```python
mean_age = np.mean(data['Age'])
print("Mean of Age is : ", mean_age)
```

```python
mean_fare = np.mean(data['Fare  '])
print("Mean of Fare is : ", mean_fare)
```

```python
mode_gender = data['Gender'].value_counts().index[0]
print("Mode of the feature 'Gender' is : ", mode_gender)
```
