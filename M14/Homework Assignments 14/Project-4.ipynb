```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/mnt/data/Company Sales Data.csv"
data = pd.read_csv(file_path)

plt.figure()
plt.plot(
    data['month_number'],
    data['total_profit'],
    linestyle=':',
    marker='o',
    linewidth=3,
    color='red',
    markerfacecolor='black'
)
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Company Profit Per Month")
plt.show()

plt.figure()
plt.plot(data['month_number'], data['facecream'], marker='o', linewidth=3, label='Face Cream')
plt.plot(data['month_number'], data['facewash'], marker='o', linewidth=3, label='Face Wash')
plt.plot(data['month_number'], data['toothpaste'], marker='o', linewidth=3, label='Toothpaste')
plt.plot(data['month_number'], data['bathingsoap'], marker='o', linewidth=3, label='Bathing Soap')
plt.plot(data['month_number'], data['shampoo'], marker='o', linewidth=3, label='Shampoo')
plt.plot(data['month_number'], data['moisturizer'], marker='o', linewidth=3, label='Moisturizer')

plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.title("Sales Data of All Products Per Month")
plt.legend()
plt.show()

plt.figure()
plt.bar(data['month_number'], data['facecream'], label='Face Cream')
plt.bar(data['month_number'], data['facewash'], bottom=data['facecream'], label='Face Wash')

plt.xlabel("Month")
```
