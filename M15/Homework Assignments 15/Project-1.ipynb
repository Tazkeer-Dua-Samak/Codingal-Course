```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
day = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
```

```python
july_births = [12, 15, 11, 9, 1, 9, 21]
august_births = [17, 5, 2, 11, 1, 8, 29]
```

```python
july = pd.DataFrame({"Day": day,"New Births": july_births})
august = pd.DataFrame({"Day": day,"New Births": august_births})
```

```python
plt.bar(july["Day"], july["New Births"], color="blue", width=0.5, edgecolor="black")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.title("Birth Rate in July")
plt.show()
```

```python
plt.bar(august["Day"], august["New Births"], color="green", width=0.5, edgecolor="black")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.title("Birth Rate in August")
plt.show()
```

```python
plt.bar(july["Day"], july["New Births"], color="blue", width=0.4, edgecolor="black", label="July")
plt.bar(august["Day"], august["New Births"], color="green", width=0.4, edgecolor="black", label="August")
plt.xlabel("Day")
plt.ylabel("New Births")
plt.title("Birth Rate Comparison of July and August")
plt.legend()
plt.show()
```
