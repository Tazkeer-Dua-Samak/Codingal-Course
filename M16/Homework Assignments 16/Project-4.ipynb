```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Bestsellers with categories.csv")
```

```python
data.info()
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
var_rating = np.var(data['User Rating'])
print("Variation in User Rating is : ", var_rating)
```

```python
std_rating = np.std(data['User Rating'])
print("Standard Deviation of User Rating is : ", std_rating)
```

```python
var_price = np.var(data['Price'])
print("Variation in Price is : ", var_price)
```

```python
std_price = np.std(data['Price'])
print("Standard Deviation of Price is : ", std_price)
```

```python
bins = np.arange(3.0, 5.1, 0.1)

plt.hist(data['User Rating'], bins=bins)
plt.ylabel("Count of Books")
plt.xlabel("User Rating")
plt.title("Distribution of User Rating")
```

```python
bins = np.arange(0, 110, 5)

plt.hist(data['Price'], bins=bins)
plt.ylabel("Count of Books")
plt.xlabel("Price")
plt.title("Distribution of Book Price")
```
