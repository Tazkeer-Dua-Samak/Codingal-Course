```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Company Sales Data.csv")
```

```python
plt.figure()
```

```python
plt.plot(data['month_number'], data['total_profit'], linestyle=':', marker='o', linewidth=3, color='red', markerfacecolor='black')
```

```python
plt.xlabel("Month")
```

```python
plt.ylabel("Profit")
```

```python
plt.title("Company Profit Per Month")
```

```python
plt.show()
```

```python
plt.figure()
```
