```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("Titanic Dataset.csv")
```

```python
data.info()
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
q1 = np.quantile(data['Age'], 0.25)
q2 = np.quantile(data['Age'], 0.50)
q3 = np.quantile(data['Age'], 0.75)
```

```python
print("First Quartile - ", q1)
```

```python
print("Second Quartile - ", q2)
```

```python
print("Third Quartile - ", q3)
```

```python
IQR_age = q3 - q1
print("Interquartile Range - ", IQR_age)
```

```python
plt.hist(data['Age'])
plt.ylabel("Count of Passengers")
plt.xlabel("Age")
```

```python
fare_q1 = np.quantile(data['Fare'], 0.25)
fare_q2 = np.quantile(data['Fare'], 0.50)
fare_q3 = np.quantile(data['Fare'], 0.75)
```

```python
print("Fare of First Quartile", fare_q1)
```

```python
print("Fare of Second Quartile", fare_q2)
```

```python
print("Fare of Third Quartile", fare_q3)
```

```python
IQR_fare = fare_q3 - fare_q1
print("Interquartile Range of Fare - ", IQR_fare)
```

```python
bins = np.arange(0,250,20)
plt.hist(data['Fare'], bins=np.arange(1,250,20))
plt.ylabel("Count of Passengers")
plt.xlabel("Fare")
plt.xticks(bins)
```
